```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math
import pandas as pd
```

## Trignometry

```python
# y=np.linspace(0.1,360,num=30)
y=np.linspace(0.1,720,num=30)

print(y)
```

```python
#sin 
y1=[ math.sin((math.pi * i)/180) for i in y]
plt.plot(y,y1)
```

```python
#cos
y2=[ math.cos((math.pi * i)/180) for i in y]
plt.plot(y,y2)
```

```python
y3=[ math.tan((math.pi * i)/180) for i in y]
plt.plot(y,y3)
```

```python
y4=[ math.sinh((math.pi * i)/180) for i in y]
plt.plot(y,y4)
```

```python
y5=[ math.cosh((math.pi * i)/180) for i in y]
plt.plot(y,y5)
```

```python
y6=[ math.tanh((math.pi * i)/180) for i in y]
plt.plot(y,y6)
```

```python
#cot
y7=[ 1/(math.tan((math.pi * i)/180)) for i in y]
plt.plot(y,y7)
```

```python
#cosec
y8=[ 1/(math.sin((math.pi * i)/180)) for i in y]
plt.plot(y,y8)
```

```python
#sec
y9=[ 1/(math.cos((math.pi * i)/180)) for i in y]
plt.plot(y,y9)
```

## Series

```python
#2n+1
X=np.arange(1,200)
y=[2*n+1 for n in X]
plt.plot(X,y)
```

```python
#log(x)
X=np.arange(1,10)
y=[np.log(x) for x in X]
plt.plot(X,y)
```

```python
#e(x)
X=np.arange(1,10)
y=[np.exp(x) for x in X]
plt.plot(X,y)
```
